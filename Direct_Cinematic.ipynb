#Basandonos en la grafica.

$\sin{\theta_{2}} = \frac{X_{2}}{A_{2}} \to X_{2}= \sin{\theta_{2}} * A_{2}= X_{2} $

$ $

$ \cos{\theta_{2}} = \frac{Y_{2}}{A_{2}} \to Y_{2} = \cos{(\theta_{2})} * A_{2}  $

$ $

$\phi{1} = \arctan{(\frac{X_{2}}{Y_{2}})} $

$ $

$\phi{2} = 180 - \theta{3} - \phi{1} $

$ $

$ \sin{\phi_{2}}= \frac{X_{3}}{A_{3}}$

$ $

$\cos{\phi_{2}} = \frac{Y_{3}}{A_{3}} $

$ $

$\sin{(\phi_{2}) * A_{3}} = X_{3}  \,\,\,\, ; \,\,\,\, \cos{\phi_{2}} * A_{3} = Y_{3} $

```python
import math

def Calculo_Variables_Cinematica(A1, A2, A3, O2 , O3):
    # Convierte los ngulos de grados a radianes si estn en grados
    O2 = math.radians(O2)
    O3 = math.radians(O3)

    # Calcula las coordenadas del segundo segmento
    X2 = math.sin(O2) * A2
    Y2 = math.cos(O2) * A2

    # Calcula el ngulo auxiliar para el tercer segmento
    ang_aux = math.atan(X2 / Y2)

    # Calcula las coordenadas del tercer segmento
    X3 = math.sin(math.pi - O3 - ang_aux) * A3
    Y3 = math.cos(math.pi - O3 - ang_aux) * A3

    return (X2, Y2, X3, Y3)
```
